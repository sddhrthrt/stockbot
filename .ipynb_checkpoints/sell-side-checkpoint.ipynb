```python
from bidder import *
import websocket
import json

account = "BMS98316411"
venue = "SOHEX"
stock = "CDS"

import sqlite3

import functools
```

```python
conn = sqlite3.connect("sell-side.db")
```

```python
conn.execute('''drop table if exists details''')
conn.execute('''create table details (account string primary key, amount int, nav int, position int)''')
conn.commit()
```

```python
def get_status(account):
    res = conn.execute('''select amount, nav, position from details where account = '%s' '''%(account,))
    row = res.fetchall()[0]
    return row
```

```python
def save_status(account, amount, nav, position):
    res = conn.execute('''insert or replace into details (account, amount, nav, position) values ('%s', %s, %s, %s)'''%(account, amount, nav, position))
    conn.commit()
```

```python
def listener(account, ws, message):
    quote = json.loads(message)['quote']
    ask, asksize, bid, bidize, venue, stock = quote.get('ask', None), quote.get('askSize', None), quote.get('bid', None), quote.get('bidSize', None), quote.get('venue'), quote.get('symbol')
    amount, nav, position = get_status(account)
    if position > 0: # sell if you can get profit
        if (amount/position) < bid:
            qty = position if (bidSize < position) else bidSize
            sell(account, venue, stock, bid, qty, 'limit')
            position -= qty
            amount += (qty * bid)
            save_status(account, amount, nav, position)
    if position < 1000: # buy for cheap
        if ask < (amount/position):
            qty = asksize if (1000-position > askSize) else 1000-position
            buy(account, venue, stock, ask, qty, 'limit')
            position += qty
            amount -= (qty * ask)
            save_status(account, amount, nav, position)
    print get_status(account)
    
```

```python
quote_websocket(account, venue, stock, functools.partial(listener, account))
```
