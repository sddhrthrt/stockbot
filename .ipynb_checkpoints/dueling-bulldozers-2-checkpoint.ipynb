```python
def quote_listener(account, ws, message):
    global amount, position, avgask, avgbid, curask, curbid, navs, nav_news, sell_orders, buy_orders
    quote = json.loads(message).get('quote', None)
    if quote is None:
        return
    myrate = abs(amount/position) if not position == 0 else 0 if not amount == 0 else avgask
    navs = navs[1:] + [amount + position * curbid]
    
    buy_orders = buy_orders[1:] + [buy_orders[-1]]
    if position < 200:
        if 'ask' in quote:
            curask = quote['ask']
            #anamoly, avgasksize = check_anamoly(avgasksize, quote['ask'])
            if quote['ask'] < myrate - 400:
                size = quote['askSize'] if (position + quote['askSize']) < 200 else 200 - position
                res = buy(account, venue, stock, quote['ask']+200, size, 'immediate-or-cancel')
                buy_oders[-1] = amount - size*(quote['ask']+200) + (position+size)*curbid
                if res['ok']:
                    amount -= res['totalFilled'] * res['price']
                    position += res['totalFilled']
                    print res['originalQty'], '/', res['totalFilled']
                else:
                    print res
        
    myrate = abs(amount/position) if not position == 0 else 0 if not amount == 0 else avgbid

    sell_orders = sell_orders[1:] + [sell_orders[-1]]
    if position > -200:
        if 'bid' in quote:
            curbid = quote['bid']
            #anamoly, avgbidsize = check_anamoly(avgbidsize, quote['bid'])
            if quote['bid'] > myrate + 400:
                size = quote['bidSize'] if (position - quote['bidSize']) > -200 else 200 + position
                res = sell(account, venue, stock, quote['bid']-200, size, 'immediate-or-cancel')
                sell_orders[-1] = amount + (quote['bid']-200)*size + (position-size)*curbid
                if res['ok']:
                    amount += res['totalFilled'] * res['price']
                    position -= res['totalFilled']
                    print res['originalQty'], '/', res['totalFilled']
                else:
                    print res
    
    if position > 200:
        if 'bid' in quote:
            if quote['bid'] > myrate + 1000:
                res = sell(account, venue, stock, quote['bid']-100, position-100, 'immediate-or-cancel')
                if res['ok']:
                    amount += res['totalFilled'] * res['price']
                    position -= res['totalfilled']
                print res
    nav_news = nav_news[1:] + [amount + position * curbid]
    
```
